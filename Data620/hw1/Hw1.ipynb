```python
#krackhardt Kite Graph
G=nx.generators.small.krackhardt_kite_graph()
```

```python
#Draw a new graph and add nodes and edges for the graph
X=nx.Graph()
X.add_node(1)
X.add_nodes_from([2,3])
```

```python
X.add_edges_from([(1, 2), (1, 3)])
nx.draw(X)
```
